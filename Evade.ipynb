```
features = data[['Refund', 'Marital_Status', 'Taxable_Income']]
features
```

```
target = data['Evade']
target
```

## 2. Apply Naive Bayes

```
from sklearn.naive_bayes import GaussianNB
from sklearn.utils.validation import column_or_1d
#Create a Gaussian Classifier
model = GaussianNB()
#Train the model
model.fit(features, target) #column_or_1d(target)
```

## 3. Check the Accuracy

```
import numpy as np
print("Prediction Accuracy is: ", model.score(features, target)*100, "%") #np.array(target)
```

```
model.classes_
```

>**X_test_1 = [['No', 'Married', 120000], ['Yes', 'Divorce', 25000]]**

```
y_pred_1 = model.predict(X_test_1)
y_pred_1
```

## 4. Save the Model

```
import pickle
pkl_evade = "evade_model.pkl" #filename
with open(pkl_evade, 'wb') as file:
  pickle.dump(model, file)
```

```
with open(pkl_evade, 'rb') as file:
  pkl_evade = pickle.load(file)
```
